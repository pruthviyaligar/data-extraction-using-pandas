## 13. Using the user defined function convert the ‘rasied_amount_usd’ column into a categorical column as follow
## a) If the amount is less than 5 Millions then recode as TypeA.
## b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
## c) If the amount is greater than 7 Millions then recode as TypeC.

```python
def usd(x):
    if x<5000000:
        return 'TYPE A'
    elif x>5000000 and x<7000000:
        return 'TYPE B'
    elif x>7000000:
        return 'TYPE C'
    else:
        return x
    
```

```python
master_dataframe['raised_amount_usd']=master_dataframe['raised_amount_usd'].apply(usd)
```
